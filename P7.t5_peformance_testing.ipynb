```python
from rouge import Rouge
import os 
from datetime import datetime
import openpyxl
import pandas as pd
import re
from sklearn.model_selection import train_test_split
```

```python
input_dir = "./Data/Generation/"
generated_column = "automatic_text"
target_column = "target_text"
results_path = "./Data/Results/{}.xlsx".format(datetime.now().strftime('%Y-%m-%d-%H-%M-%S'))
```

```python
model_name_lst = []
rouge_1_p_lst = []
rouge_1_r_lst = []
rouge_1_f_lst = []
rouge_2_p_lst = []
rouge_2_r_lst = []
rouge_2_f_lst = []
rouge_l_p_lst = []
rouge_l_r_lst = []
rouge_l_f_lst = []

rouge_1_p_lst_raw = []
rouge_1_r_lst_raw = []
rouge_1_f_lst_raw = []
rouge_2_p_lst_raw = []
rouge_2_r_lst_raw = []
rouge_2_f_lst_raw = []
rouge_l_p_lst_raw = []
rouge_l_r_lst_raw = []
rouge_l_f_lst_raw = []
bert_f_score_lst = []
bert_precision_lst = []
bert_recall_lst = []

rouge = Rouge()
for df in os.listdir(input_dir):
    if (df != ".ipynb_checkpoints"): 
        
        print(df)
       
        data = pd.read_pickle(input_dir + df)
        
        if  type(data[generated_column].to_list()[0]) == str:
            automatic = data[generated_column].to_list()
        else:
            automatic = [sentence[0] for sentence in data[generated_column].to_list()]
            
        gold = data[target_column].to_list()
        scores = rouge.get_scores(automatic , gold, avg=True)
        #P, R, F1 = score(automatic, gold, lang="cs", verbose=True)
        #rouge_raw = RougeRaw().corpus(gold, automatic)


        model_name_lst.append(df)
        rouge_1_p_lst.append(scores["rouge-1"]["p"])
        rouge_1_r_lst.append(scores["rouge-1"]["r"])
        rouge_1_f_lst.append(scores["rouge-1"]["f"])
        rouge_2_p_lst.append(scores["rouge-2"]["p"])
        rouge_2_r_lst.append(scores["rouge-2"]["r"])
        rouge_2_f_lst.append(scores["rouge-2"]["f"])
        rouge_l_p_lst.append(scores["rouge-l"]["p"])
        rouge_l_r_lst.append(scores["rouge-l"]["r"])
        rouge_l_f_lst.append(scores["rouge-l"]["f"])

        """rouge_1_p_lst_raw.append(rouge_raw['1'].p * 100)
        rouge_1_r_lst_raw.append(rouge_raw['1'].r * 100)
        rouge_1_f_lst_raw.append(rouge_raw['1'].f * 100)
        rouge_2_p_lst_raw.append(rouge_raw['2'].p * 100)
        rouge_2_r_lst_raw.append(rouge_raw['2'].r * 100)
        rouge_2_f_lst_raw.append(rouge_raw['2'].f * 100)
        rouge_l_p_lst_raw.append(rouge_raw['L'].p * 100)
        rouge_l_r_lst_raw.append(rouge_raw['L'].r * 100)
        rouge_l_f_lst_raw.append(rouge_raw['L'].f * 100)"""

        #bert_f_score_lst.append(float(torch.mean(F1)))
        #bert_precision_lst.append(float(torch.mean(P)))
        #bert_recall_lst.append(float(torch.mean(R)))

        #float(torch.mean(P))
    
result_df = pd.DataFrame(
            {    
                'model': model_name_lst,
                'rouge-1_f_score': rouge_1_f_lst,
                'rouge-2_f_score': rouge_2_f_lst,
                'rouge-l_f_score':  rouge_l_f_lst,
                #'rougeRAW-1_f_score': rouge_1_f_lst_raw,
                #'rougeRAW-2_f_score': rouge_2_f_lst_raw,
                #'rougeRAW-l_f_score':  rouge_l_f_lst_raw,
                #'BERTScore_f_score': bert_f_score_lst,
                
                'rouge-1_precision': rouge_1_p_lst,
                'rouge-2_precision': rouge_2_p_lst,
                'rouge-l_precision': rouge_l_p_lst,
                #'rougeRAW-1_precision': rouge_1_p_lst_raw,
                #'rougeRAW-2_precision': rouge_2_p_lst_raw,
                #'rougeRAW-l_precision': rouge_l_p_lst_raw,
                #'BERTScore_precision': bert_precision_lst,
                
                'rouge-1_recall':  rouge_1_r_lst,
                'rouge-2_recall': rouge_2_r_lst,
                'rouge-l_recall':  rouge_l_r_lst,
                #'rougeRAW-1_recall':  rouge_1_r_lst_raw,
                #'rougeRAW-2_recall': rouge_2_r_lst_raw,
                #'rougeRAW-l_recall':  rouge_l_r_lst_raw,
                #'BERTScore_recall': bert_recall_lst,
                
            }
      )
result_df.to_excel(results_path)
```
